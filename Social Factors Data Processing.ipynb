```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
%matplotlib inline
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
mp = pd.read_csv("SeattleData/MarketProfile/R13142629_SL150.csv")
```

```python
mp.head()
```

### Travel to work

```python
ttw = mp[['Geo_FIPS','ORG_TRAV30_59','ORG_TRAV60_89','ORG_TRAV90P','ORG_TRANCAR','ORG_TRANPUBLIC','ORG_TRANBCYCL','ORG_TRANWALK','ORG_TOT_POP10']]
ttw.head()
```

```python
ttw['ttw_30_59%']  = ttw['ORG_TRAV30_59']/ttw['ORG_TOT_POP10']
ttw['ttw_60_89%']  = ttw['ORG_TRAV60_89']/ttw['ORG_TOT_POP10']
ttw['ttw_90%']  = ttw['ORG_TRAV90P']/ttw['ORG_TOT_POP10']
ttw['ttw_bycar%'] = ttw['ORG_TRANCAR']/ttw['ORG_TOT_POP10']
ttw['ttw_bytransit%'] =ttw['ORG_TRANPUBLIC']/ttw['ORG_TOT_POP10']
ttw['ttw_bybike%'] =ttw['ORG_TRANBCYCL']/ttw['ORG_TOT_POP10']
ttw['ttw_byfoot%'] =ttw['ORG_TRANWALK']/ttw['ORG_TOT_POP10']
ttw.head()
```

```python
ttw_summary = ttw[['Geo_FIPS','ttw_30_59%','ttw_60_89%','ttw_90%','ttw_bycar%','ttw_bytransit%','ttw_bybike%','ttw_byfoot%']]
ttw_summary.head()
```

### Median Rent

```python
rent = mp[['Geo_FIPS','ORG_MEDRENT']]
rent = rent.rename(columns={'ORG_MEDRENT': 'median_rent'})
```

```python
rent.head()
```

### Income

```python
# household income median, per capital income
income = mp[['Geo_FIPS','ORG_MEDHHINC','ORG_PERCAPINC']]
income = income.rename(columns={'ORG_MEDHHINC': 'income_household_median','ORG_PERCAPINC': 'income_percapita'})
income.head()
```

### Quality of Life

https://www.easidemographics.com/mdbhelp/html/quality_of_life_1.htm

```python
qol = mp[['Geo_FIPS','ORG_EASIQLIFE']]
qol = qol.rename(columns={'ORG_EASIQLIFE': 'quality_of_life_score'})
qol.head()
```

### Consumer Expenditure

```python
expenditure = mp[['Geo_FIPS','ORG_ANNEXP','ORG_TRANSPORT']]
```

```python
expenditure['transport_expenditure%']=expenditure['ORG_TRANSPORT']/expenditure['ORG_ANNEXP']
expenditure.head()
```

```python
expenditure_summary =expenditure [['Geo_FIPS','transport_expenditure%']]
expenditure_summary.head()
```

### Population Density

population density 2020

```python
density = mp[['Geo_FIPS','ORG_TOTPOP20201']]
density = density.rename(columns={'ORG_TOTPOP20201': 'ppl_density'})
density.head()
```

```python
df = ttw_summary.merge(rent, on='Geo_FIPS').merge(income, on='Geo_FIPS').merge(qol, on='Geo_FIPS').merge(expenditure_summary, on='Geo_FIPS').merge(density, on='Geo_FIPS')
```

```python
df.head()
```

```python
df.to_csv('SeattleData/SeattleSocialFactors.csv',index=False)
```
