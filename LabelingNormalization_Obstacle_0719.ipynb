```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
%matplotlib inline
import seaborn as sns; sns.set_theme(color_codes=True)
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors
import warnings
warnings.filterwarnings('ignore')
from scipy import stats
```

```python
sidewalk = gpd.read_file('SeattleData/ProjectSidewalkLabelsMappedonSidewalk/ProjectSidewalkLabelsMappedonSidewalk.shp')
sidewalk.head()
```

```python
labels = gpd.read_file('SeattleData/ProjectSidewalkData/ProjectSidewalkData_Geolabels.shp')
labels.head()
```

```python
#Spatial join project sidewalk labels to sidewalk geometry
labels_mapped_on_sidewalk = labels.sjoin_nearest(sidewalk)
labels_mapped_on_sidewalk.head()
```

```python
#Select label types : 'Obstacle'
Obstacle = labels_mapped_on_sidewalk.loc[(labels_mapped_on_sidewalk['Label Type'] == 'Obstacle')] 
Obstacle.head()
```

```python
#Cleaning data
Obstacle = Obstacle.loc[Obstacle['Disagree C'] == 0] #Takeout ones with disagreecounts
Obstacle = Obstacle.dropna(subset=['Severity']) #Takeout Severity NA datas
Obstacle = Obstacle.rename(columns={'Severity': 'WeightedCount'})
```

```python
#Group label data by street ID
Obstacle_grouped = Obstacle.groupby(by=["OBJECTID"]).sum()
Obstacle_grouped.head()
```

```python
#Merge back to sidewalk geometry
sidewalk_w_labels = sidewalk.merge(Obstacle_grouped, how='outer', on='OBJECTID').fillna(0)
sidewalk_w_labels.head()
```

```python
#Select only useful columns
Obstacle_per_street = sidewalk_w_labels[['OBJECTID','CONDITION','WeightedCount','SHAPE_Leng_x','geometry']]
Obstacle_per_street.head()
```

```python
#Normalize Obstacle count by street length
Obstacle_per_street ['Obstacle_WeightedCount_per_feet'] = (Obstacle_per_street ['WeightedCount']/Obstacle_per_street ['SHAPE_Leng_x'])
Obstacle_per_street.head()
```

```python
#Check boxplot
sns.histplot(data=Obstacle_per_street, x="Obstacle_WeightedCount_per_feet")
```

```python
#Remove outliers
q = Obstacle_per_street["Obstacle_WeightedCount_per_feet"].quantile(0.99)
Obstacle_per_street_n = Obstacle_per_street[Obstacle_per_street["Obstacle_WeightedCount_per_feet"] < q]
```

```python
#Group to block data
bg = gpd.read_file('SeattleData/Census_Block_Groups_2010/Census_Block_Groups_2010.shp')
bg.head()
```

```python
sidewalk_per_bg=Obstacle_per_street_n.sjoin(bg, op="within")
sidewalk_per_bg_grouped = sidewalk_per_bg.groupby(by=["GEOID10"]).sum()
sidewalk_per_bg_grouped.head()
```

```python
#Normalize sidewalk Obstacle count in each block group
sidewalk_per_bg_grouped= sidewalk_per_bg_grouped[['WeightedCount','SHAPE_Leng_x']].reset_index().rename(columns={'SHAPE_Leng_x': 'SidewalkLeng'})
sidewalk_per_bg_grouped['Obstacle_WeightedCount_per_mile'] = (sidewalk_per_bg_grouped['WeightedCount']/sidewalk_per_bg_grouped['SidewalkLeng'])* 5280
sidewalk_per_bg_grouped.head()
```

```python
#Check boxplot
sns.histplot(data=sidewalk_per_bg_grouped, x="Obstacle_WeightedCount_per_mile")
```

```python
sidewalk_per_bg_grouped.to_csv('SeattleData/SeattleNormalizedSidewalkLabelCounts_Obstacle_0720.csv',index=False)
```
