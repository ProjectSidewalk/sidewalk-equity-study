```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
%matplotlib inline
import seaborn as sns; sns.set_theme(color_codes=True)
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors
import warnings
warnings.filterwarnings('ignore')
from scipy import stats
```

```python
#Get sidewalk geo data
s = gpd.read_file('SeattleData/Sidewalks/Sidewalks.shp')
s
```

```python
s_clean = s.dropna(subset=['geometry'])
s_clean = s_clean.drop_duplicates(subset=['geometry'])
s_clean
```

```python
#project sidewalk data
labels = pd.read_csv("SeattleData/access_attributes.csv")
labels.head()
```

```python
#create labels geo dataframe
labels = labels.rename(columns={'Attribute Latitude': 'latitude', 'Attribute Longitude': 'longtitude'})
points = labels.apply(lambda row: Point(row.longtitude, row.latitude), axis=1)
labels_geo = gpd.GeoDataFrame(labels, geometry=points)
labels_geo.crs = {'init': 'epsg:4326'}
labels_geo = labels_geo.to_crs(4326)
labels_geo.head()
```

```python
labels_geo.to_file('SeattleData/ProjectSidewalkData/ProjectSidewalkData_Geolabels.shp') 
```

```python
#Spatial join project sidewalk labels to sidewalk geometry
labels_per_street = labels_geo.sjoin_nearest(s_clean)
```

```python
labels_per_street.head()
```

```python
street_w_labels = labels_per_street.groupby(by=["OBJECTID"]).sum()
```

```python
street_w_labels
```

```python
street_w_labels= street_w_labels['Severity'].reset_index().rename(columns={'Severity': 'Problems_WeightedCount'})
```

```python
street_w_labels
```

```python
sidewalk_w_labels = s_clean.merge(street_w_labels, how='inner', on='OBJECTID')
```

```python
sidewalk_w_labels.head()
```

```python
sidewalk_w_labels = sidewalk_w_labels.drop(columns=['Problems_WeightedCount'])
```

```python
sidewalk_w_labels
```

```python
sidewalk_w_labels.to_file('SeattleData/ProjectSidewalkLabelsMappedonSidewalk/ProjectSidewalkLabelsMappedonSidewalk.shp') 
```
