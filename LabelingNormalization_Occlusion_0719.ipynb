```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
%matplotlib inline
import seaborn as sns; sns.set_theme(color_codes=True)
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors
import warnings
warnings.filterwarnings('ignore')
from scipy import stats
```

```python
sidewalk = gpd.read_file('SeattleData/ProjectSidewalkLabelsMappedonSidewalk/ProjectSidewalkLabelsMappedonSidewalk.shp')
sidewalk.head()
```

```python
labels = gpd.read_file('SeattleData/ProjectSidewalkData/ProjectSidewalkData_Geolabels.shp')
labels.head()
```

```python
df =labels[labels['Label Type']=='Occlusion']
```

```python
df['Severity'].value_counts(dropna=False)
```

```python
#Spatial join project sidewalk labels to sidewalk geometry
labels_mapped_on_sidewalk = labels.sjoin_nearest(sidewalk)
labels_mapped_on_sidewalk.head()
```

```python
#Select label types : 'Occlusion'
Occlusion = labels_mapped_on_sidewalk.loc[(labels_mapped_on_sidewalk['Label Type'] == 'Occlusion')] 
Occlusion
```

```python
Occlusion['Severity'].value_counts(dropna=False)
```

```python
#Cleaning data
Occlusion = Occlusion.loc[Occlusion['Disagree C'] == 0] #Takeout ones with disagreecounts
Occlusion = Occlusion.dropna(subset=['Severity']) #Takeout Severity NA datas
Occlusion = Occlusion.rename(columns={'Severity': 'WeightedCount'})
```

```python
#Group label data by street ID
Occlusion_grouped = Occlusion.groupby(by=["OBJECTID"]).sum()
Occlusion_grouped.head()
```

```python
#Merge back to sidewalk geometry
sidewalk_w_labels = sidewalk.merge(Occlusion_grouped, how='outer', on='OBJECTID').fillna(0)
sidewalk_w_labels.head()
```

```python
#Select only useful columns
Occlusion_per_street = sidewalk_w_labels[['OBJECTID','CONDITION','WeightedCount','SHAPE_Leng_x','geometry']]
Occlusion_per_street.head()
```

```python
#Normalize Occlusion count by street length
Occlusion_per_street ['Occlusion_WeightedCount_per_feet'] = (Occlusion_per_street ['WeightedCount']/Occlusion_per_street ['SHAPE_Leng_x'])
Occlusion_per_street.head()
```

```python
#Check boxplot
sns.histplot(data=Occlusion_per_street, x="Occlusion_WeightedCount_per_feet")
```

```python
#Remove outliers
q = Occlusion_per_street["Occlusion_WeightedCount_per_feet"].quantile(0.99)
Occlusion_per_street_n = Occlusion_per_street[Occlusion_per_street["Occlusion_WeightedCount_per_feet"] < q]
```

```python
#Group to block data
bg = gpd.read_file('SeattleData/Census_Block_Groups_2010/Census_Block_Groups_2010.shp')
bg.head()
```

```python
sidewalk_per_bg=Occlusion_per_street_n.sjoin(bg, op="within")
sidewalk_per_bg_grouped = sidewalk_per_bg.groupby(by=["GEOID10"]).sum()
sidewalk_per_bg_grouped.head()
```

```python
#Normalize sidewalk Occlusion count in each block group
sidewalk_per_bg_grouped= sidewalk_per_bg_grouped[['WeightedCount','SHAPE_Leng_x']].reset_index().rename(columns={'SHAPE_Leng_x': 'SidewalkLeng'})
sidewalk_per_bg_grouped['Occlusion_WeightedCount_per_mile'] = (sidewalk_per_bg_grouped['WeightedCount']/sidewalk_per_bg_grouped['SidewalkLeng'])* 5280
sidewalk_per_bg_grouped.head()
```

```python
#Check boxplot
sns.histplot(data=sidewalk_per_bg_grouped, x="Occlusion_WeightedCount_per_mile")
```

```python
sidewalk_per_bg_grouped.to_csv('SeattleData/SeattleNormalizedSidewalkLabelCounts_Occlusion_0720.csv',index=True)
```
