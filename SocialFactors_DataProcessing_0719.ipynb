```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
%matplotlib inline
import seaborn as sns
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
```

```python
mp = pd.read_csv("SeattleData/MarketProfile/R13142629_SL150.csv")
```

```python
#mp
```

## Family Population

```python
family_ppl = mp[['Geo_FIPS','ORG_TOT_POP0','ORG_POPFAM']]
family_ppl['family_ppl%']  = family_ppl['ORG_POPFAM']/family_ppl['ORG_TOT_POP0']
family_ppl = family_ppl[['Geo_FIPS','family_ppl%']]
family_ppl.head()
```

## Travel to work

```python
ttw = mp[['Geo_FIPS','ORG_TRAV30_59','ORG_TRAV60_89','ORG_TRAV90P','ORG_TRANCAR','ORG_TRANPUBLIC','ORG_TRANBCYCL','ORG_TRANWALK','ORG_TOT_POP10']]
ttw.head()
```

```python
ttw['ttw_30_59%']  = ttw['ORG_TRAV30_59']/ttw['ORG_TOT_POP10']
ttw['ttw_60_89%']  = ttw['ORG_TRAV60_89']/ttw['ORG_TOT_POP10']
ttw['ttw_90%']  = ttw['ORG_TRAV90P']/ttw['ORG_TOT_POP10']
ttw['ttw_bycar%'] = ttw['ORG_TRANCAR']/ttw['ORG_TOT_POP10']
ttw['ttw_bytransit%'] =ttw['ORG_TRANPUBLIC']/ttw['ORG_TOT_POP10']
ttw['ttw_bybike%'] =ttw['ORG_TRANBCYCL']/ttw['ORG_TOT_POP10']
ttw['ttw_byfoot%'] =ttw['ORG_TRANWALK']/ttw['ORG_TOT_POP10']
ttw.head()
```

```python
ttw_summary = ttw[['Geo_FIPS','ttw_30_59%','ttw_60_89%','ttw_90%','ttw_bycar%','ttw_bytransit%','ttw_bybike%','ttw_byfoot%']]
ttw_summary.head()
```

## Median Vehicles per Household

```python
median_vehicles = mp[['Geo_FIPS','ORG_MEDVEHICLE']]
median_vehicles = median_vehicles.rename(columns={'ORG_MEDVEHICLE': 'median_vehicles_per_household'})
median_vehicles.head()
```

## Population Change

```python
ppl_growth = mp[['Geo_FIPS','ORG_POPGROW20','ORG_POPFORE25']]
ppl_growth = ppl_growth.rename(columns={'ORG_POPGROW20': 'ppl_growth%_2010_2020','ORG_POPFORE25': 'ppl_growth%_2020_2025'})
ppl_growth.head()
```

## Median Rent

## Income

## Daytime Population

```python
day_ppl = mp[['Geo_FIPS','ORG_EASIDAYPOP']]
day_ppl = day_ppl.rename(columns={'ORG_EASIDAYPOP': 'daytime_ppl'})
day_ppl.head()
```

## Consumer Expenditure

### Consumer Expenditures, Transportation

```python
mp_te = pd.read_csv("SeattleData/MarketProfile/2021/Transportation/R13149034_SL150.csv")
```

```python
transportation_expenditure = mp_te[['Geo_FIPS','ORG_TRANSPORT1','ORG_VEHPURCH','ORG_GAS','ORG_OTHRVEH','ORG_PUBTRANS']]
```

```python
transportation_expenditure['vehicle_purchase%']  = transportation_expenditure['ORG_VEHPURCH']/transportation_expenditure['ORG_TRANSPORT1']
transportation_expenditure['gas%']  = transportation_expenditure['ORG_GAS']/transportation_expenditure['ORG_TRANSPORT1']
transportation_expenditure['vehicle_others%']  = transportation_expenditure['ORG_OTHRVEH']/transportation_expenditure['ORG_TRANSPORT1']
transportation_expenditure['public_transport%']  = transportation_expenditure['ORG_PUBTRANS']/transportation_expenditure['ORG_TRANSPORT1']
```

```python
transportation_expenditure = transportation_expenditure[['Geo_FIPS','vehicle_purchase%','gas%','vehicle_others%','public_transport%']]
```

```python
transportation_expenditure.head()
```

## Housing Units In Structure

```python
mp_2021 = pd.read_csv("SeattleData/MarketProfile/2021/HousingStructure/R13148971_SL150.csv")
```

```python
housing = mp_2021[['Geo_FIPS','ORG_OCCUNIT2','ORG_S_1D','ORG_S_1A','ORG_S_2','ORG_S_3_4','ORG_S_5_9','ORG_S_10_19','ORG_S_20_49','ORG_S_50P']]
```

```python
housing['hs_1_detached%']  = housing['ORG_S_1D']/housing['ORG_OCCUNIT2']
housing['hs_1_attached%']  = housing['ORG_S_1A']/housing['ORG_OCCUNIT2']
housing['hs_2%']  = housing['ORG_S_2']/housing['ORG_OCCUNIT2']
housing['hs_3_4%']  = housing['ORG_S_3_4']/housing['ORG_OCCUNIT2']
housing['hs_5_9%']  = housing['ORG_S_5_9']/housing['ORG_OCCUNIT2']
housing['hs_10_19%']  = housing['ORG_S_10_19']/housing['ORG_OCCUNIT2']
housing['hs_20_49%']  = housing['ORG_S_20_49']/housing['ORG_OCCUNIT2']
housing['hs_50p%']  = housing['ORG_S_50P']/housing['ORG_OCCUNIT2']
```

```python
housing = housing [['Geo_FIPS','hs_1_detached%','hs_1_attached%','hs_2%','hs_3_4%','hs_5_9%','hs_10_19%','hs_20_49%','hs_50p%']]
```

```python
df = family_ppl.merge(ttw_summary, on='Geo_FIPS').merge(median_vehicles, on='Geo_FIPS').merge(ppl_growth, on='Geo_FIPS').merge(rent, on='Geo_FIPS').merge(income, on='Geo_FIPS').merge(qol, on='Geo_FIPS').merge(day_ppl, on='Geo_FIPS').merge(expenditure_summary, on='Geo_FIPS').merge(transportation_expenditure, on='Geo_FIPS').merge(density, on='Geo_FIPS').merge(housing, on='Geo_FIPS').merge(mortality, on='Geo_FIPS').merge(crime, on='Geo_FIPS')
```
